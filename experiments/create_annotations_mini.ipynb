```python
import os
import matplotlib.pyplot as plt
import pandas as pd
```

```python
os.chdir('../scripts')
from utils import create_annotations_mini
os.chdir('../experiments')
```

```python
annotations_path = '../data/annotations/annotations.xlsx'
annotations_mini_path = '../data/annotations/annotations_mini.xlsx'
```

```python
odir_df = pd.read_excel(annotations_path)
```

```python
label_names = ['A','C','D','G','H','M','N','O']

plt.figure(figsize=(27,3))
for i, label in enumerate(label_names):
    total_label = int(odir_df.loc[:, [label]].sum())

    plt.subplot(1,8,i+1)
    odir_df.pivot_table(index=label, aggfunc='size').plot(kind='bar', title = f'Train class distribution ({total_label})')
```

```python
odir_mini_df = create_annotations_mini(source_path = annotations_path,
                                       target_path = annotations_mini_path,
                                       labels = ['D', 'M', 'N'],
                                       images_per_label = 100)
odir_mini_df
```

```python
label_names = ['D', 'M', 'N']

plt.figure(figsize=(9,3))
for i, label in enumerate(label_names):
    total_label = int(odir_mini_df.loc[:, [label]].sum())

    plt.subplot(1,3,i+1)
    odir_mini_df.pivot_table(index=label, aggfunc='size').plot(kind='bar', title = f'Train class distribution ({total_label})')
```
